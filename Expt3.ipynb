```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

LINE cHART

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['sepal_length'], label='Sepal Length', marker='o')
plt.plot(df.index, df['sepal_width'], label='Sepal Width', marker='x')
plt.title('Sepal Length and Width over Samples')
plt.xlabel('Sample Index')
plt.ylabel('Measurements (cm)')
plt.legend()
plt.show()
```

COUNTPLOT

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df)
plt.title('Count of Each Species in the Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
import matplotlib.pyplot as plt
species_counts = df['species'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Iris Species')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
plt.hist(df['sepal_length'], bins=10, edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```
