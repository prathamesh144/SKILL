```python
data1=pd.DataFrame(data1)
data2=pd.DataFrame(data2)
```

```python
# prompt: merge these 2  datasets using merge function

merged_data = pd.merge(data1, data2, on='longitude', how='inner')
merged_data.head()
```

```python
# prompt: concatenate these 2 datasets

concatenated_data = pd.concat([data1, data2], ignore_index=True)
concatenated_data.head()
```
