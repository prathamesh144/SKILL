```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

1. Label Encoding

```python
data['title'].fillna('Unknown', inplace=True)
data['genres'].fillna('Unknown', inplace=True)
data['releaseYear'].fillna(data['releaseYear'].median(), inplace=True)
data['imdbAverageRating'].fillna(data['imdbAverageRating'].median(), inplace=True)
data['imdbNumVotes'].fillna(data['imdbNumVotes'].median(), inplace=True)
```

2. Label Encoding for type

```python
label_encoder = LabelEncoder()
data['type'] = label_encoder.fit_transform(data['type'])
```

One Hot Encoding

```python
top_genres = data['genres'].str.split(',').explode().value_counts().head(5).index
for genre in top_genres:
    data[f'genre_{genre}'] = data['genres'].apply(lambda x: 1 if genre in x else 0)
```

Frequency Encoding

```python
data['country_count'] = data['availableCountries'].apply(lambda x: len(str(x).split(',')))
```
